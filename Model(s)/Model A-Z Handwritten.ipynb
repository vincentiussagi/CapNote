```python
x = frame.drop('0', axis = 1)
y = frame['0']
```

```python
# Reshaping the data so that they can be displayed as an image, 
# In CSV : 784 column --> convert it to 28×28 pixels.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
x_train = np.reshape(x_train.values, (x_train.shape[0], 28, 28))
x_test = np.reshape(x_test.values, (x_test.shape[0], 28, 28))

print('Train Data Shape:', x_train.shape)
print('Test Data Shape:', x_test.shape)
```

```python
# create a dictionary word_dict to map the integer values with the characters.

word_dict = {
    0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J',10:'K',11:'L',12:'M',13:'N',14:'O',15:'P',16:'Q',17:'R',18:'S',19:'T',20:'U',21:'V',22:'W',23:'X', 24:'Y',25:'Z'
}
```

```python
plt.style.use('bmh')

# Convert label into integer
y_integer = np.int0(y)
count = np.zeros(26, dtype = 'int')

# append into the count list according to the label
for i in y_integer:
    count[i] += 1

# Create a list containing all characters  
alphabets = []

for i in word_dict.values():
    alphabets.append(i)
    
fig, ax = plt.subplots(1, 1, figsize = (15, 15))
ax.bar(alphabets, count)

plt.ylabel('Number Of Elements', fontsize = 20, fontweight = 'bold', color = 'green')
plt.xlabel('Alphabets', fontsize = 30, fontweight = 'bold', color = 'green')
plt.grid()
plt.show()
```

```python
# Reshaping the training & test dataset so that it can be put in the model

x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 1)

print("New shape of train data:", x_train.shape)
print("New shape of test data:", x_test.shape)
```

```python
categorical_train = to_categorical(y_train, num_classes = 26, dtype = 'int')
print("New shape of train labels:", categorical_train.shape)

categorical_test = to_categorical(y_test, num_classes = 26, dtype = 'int')
print("New shape of test labels:", categorical_test.shape)
```

```python
model = Sequential([
    Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)),
    MaxPool2D(pool_size = (2, 2), strides = 2),
    Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu', padding = 'same'),
    MaxPool2D(pool_size = (2, 2), strides = 2),
    Conv2D(filters = 128, kernel_size = (3, 3), activation = 'relu', padding = 'valid'),
    MaxPool2D(pool_size = (2, 2), strides = 2),
    Flatten(),
    Dense(64, activation = "relu"),
    Dense(128, activation = "relu"),
    Dense(26, activation = "softmax")
])
```

```python
import tensorflow as tf

class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get("accuracy") > 0.97 and logs.get("val_accuracy") > 0.97:
            print("\n** Stopping training because accuracy has reached 97%")
            self.model.stop_training = True

callback=MyCallback()
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
history = model.fit(x_train, categorical_train, epochs = 5, validation_data = (x_test, categorical_test),callbacks=callback)
```

```python
model.summary()
model.save(r'model_hand.h5')
```

```python
print("The validation accuracy is :", history.history['val_accuracy'])
print("The training accuracy is :", history.history['accuracy'])
print("The validation loss is :", history.history['val_loss'])
print("The training loss is :", history.history['loss'])
```
